```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
d=pd.read_csv('Sales.csv', encoding= 'unicode_escape')
#unicode_escape is used to handle errors that are occured due to decoding(some special charcaters in ) csv file with utf-8 
```

```python
d.shape
```

```python
d.head(20)
```

```python
#cleaning  the data
# removing the blank values/columns
d.drop('unnamed1',axis=1,inplace=True)
#inplace=True is used to save the changes that we have done in data
#if we did'nt use true then the changes won't save permantly 
d.drop('Status',axis=1,inplace=True)
```

```python
#Finding any null values

d.isnull().sum()
```

```python
#removing any null values
d.dropna(inplace=True)
```

```python
d['Amount'].dtypes
```

```python
d.isnull().sum()
```

```python
#converting amount to int
d['Amount'] = d['Amount'].astype('int')
```

```python
d.describe()
```

```python
d.info()
```

```python
d.columns
```

```python
#Main thing: let's find the male and female count
a=sns.countplot(x='Gender',data=d)
for i in a.containers:
    a.bar_label(i)
```

```python
#plotting the graph for amount and gender

sales_gen = d.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

```python
#from the above clearly female has highest spends than men
#hue is used in plotting using seaborn it is used to plot different types of categories in same domain
#like see in the age group there are many age groups, using hue we can able to plot the graph
a=sns.countplot(data=d,x='Age Group',hue='Gender')
for bars in a.containers:
    a.bar_label(bars)
```

```python
#from the above graph clearly age(26-35) group spends alot in diwali season especially Females

sales_state = d.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)


sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
#see in above graph you can see the names are like stricking each other,
#in order to avoid that there is a function called 'rc' it is used to set the size of the graph
#if we didn't mention, then it'll plot default size


#total number of 'orders' from top 10 states
sns.set(rc={'figure.figsize':(17,7)})
plt.title("Top orders from the data")

sns.barplot(data = sales_state, x = 'State',y= 'Orders')
plt.show()
```

```python
#Now let's see the spending in state wise order
sales_state = d.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,7)})
plt.title("Top Amount from the data")

sns.barplot(data = sales_state, x = 'State',y= 'Amount')
plt.show()
```

```python
a=sns.countplot(data=d,x='Marital_Status')
#setting the size for plotting a graph
sns.set(rc={'figure.figsize':(7,7)})
plt.legend(['unmarried','married'])
for i in a.containers:
    a.bar_label(i)
plt.show()
```

```python
#Marital Status=0(Unmarried) else Married
#from this un married people is spending.

#Now let's gender in married and unmarried 
sales_state = d.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
#hue is used in plotting using seaborn it is used to plot different types of categories in same domain
```

```python
#from above graph we can conclude that unmarried female spends more than married female 
#Also, unmarried male spends morethan married male


# let's see their occupations who spends more
sales_state = d.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

```python
#let's see the Top 10 products that most of people spends
sales_state = d.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

```python
'''most of people spends more on food, clothing, gadgets,etc...
#UnMarried women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category'''
```
